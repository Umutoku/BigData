```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
.master("local[4]") \
.appName("DF to Disk") \
.config("spark.executor.memory","4g") \
.config("spark.driver.memory","2g") \
.getOrCreate()

adultDataDf = spark.read \
.option("header","True") \
.option("inferSchema","True") \
.option("sep",",") \
.csv("D:\\datasets\\adult.data")
```

```python
adultTestDf = spark.read \
.option("header","True") \
.option("inferSchema","True") \
.option("sep",",") \
.csv("D:\\datasets\\adult.test")
```

```python
adultTestDf.show(10)
```

```python
adultDataDf.show(10)
```

```python
adultWholeDf = adultDataDf.union(adultTestDf)
```

```python
adultWholeDf.show(5)
```

```python
from pyspark.sql import functions as F
adultWholeDf.groupBy(F.col("workclass")) \
.agg({"*":"count"}) \
.toPandas().head(10)
```
